# U.S. Medical Insurance Costs

### Import, see and analyse the data:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("insurance.csv")

df = df.drop_duplicates()

print(df.dtypes)
```

### Factors that influence the charges

Histogram of the costs

```python
plt.hist(df.charges, bins=10)
plt.show()
```

If we are analysing the costs of the insurance, it is helpfull to understand which variables are more correlated with the charges.

```python
# Define the colormap which maps the data values to the color space defined with the diverging_palette method  
colors = sns.diverging_palette(150, 275, s=80, l=55, n=9, as_cmap=True)
 
# Create heatmap using the .corr method on df, set colormap to cmap
sns.heatmap(df.corr(), center=0, cmap=colors, robust=True)
plt.show()
```

From the heatmap we can conclude that the charges are more positively correlated with the age, followed by the bmi, and finally the number of children.

- Plot age vs charges

```python
sns.scatterplot(x="age", y="charges", data=df)
plt.show()
```

It is clear that as the age increases, the costs increase. However, it appears to be divided in three groups: <1000, 1000<x<2500 and >3000. Probably because the age is not the only factor influencing the cost.

- Plot sex vs charges

```python
sns.boxplot(x="sex", y="charges", data=df)
plt.show()
```

Although more males are paying higher prices, the boxplots overlap meaning that probably the sex does not influence the charges.

- Plot bmi vs charges

```python
sns.scatterplot(x="bmi", y="charges", data = df)
plt.show()
```

For bmi<30, the charges are always <40000. For bmi>30 we can either have charges <40000 or >40000 showing that we might have two groups. The bmi does not seem to affect the charges in an obvious way.

- Plot children vs charges

```python
sns.scatterplot(x="children", y="charges", data = df)
plt.show()
```

The number of children does not affect the charges except when the number equals 4, or 5. Probably for big families there is a discount in the insurance.

- Plot smoker vs charges

```python
sns.boxplot(x="smoker", y="charges", data=df)
plt.show()
```

Clearly the smokers pay a higher insurance cost.

- Plot region vs charges

```python
sns.boxplot(x="region", y="charges", data=df)
plt.show()
```

Since the boxplots overlap, the region probably does not influence the charges.

**Conclusion** - The variable that higly influences the charges is "smoker". If a person smokes, the price of the insurance is higher. Maybe this variable can explain the differences in the graphs age vs charges and bmi vs charges.

```python
#Create again the plots with hue smoker
plt.subplot(1,2,1)
sns.scatterplot(x="age", y="charges", data=df, hue="smoker")
plt.title("Age vs charges")

plt.subplot(1,2,2)
sns.scatterplot(x="bmi", y="charges", data = df, hue="smoker")
plt.title("bmi vs charges")
plt.subplots_adjust(right=3)
plt.show()
```

**Final thoughts:** 
- The cost of the insurance tends to increase with age, which makes sense because an older person tends to have more health problems;
- For the same age, it is clear that a smoker pays a higher insurance cost.
- People with bmi>30, smokers, pay the highest prices for the insurance (>30000). People with bmi<30, if smokers pay a higher price commpared to non-smokers, but it is always <40000.
- The bmi (body mass index) measures "how healthy is our weight". People with a bmi>30 are considered not healthy, smoking is an additional risk factor, that explains the high insurance costs.
